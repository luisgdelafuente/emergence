```python
import numpy as np
import matplotlib.pyplot as plt

def SIR_model(N, beta, gamma, mu, I0, R0, days):
    # Initialize lists to store the results
    S = [N - I0 - R0]
    I = [I0]
    R = [R0]
    D = [0]  # List to store the deceased

    # Simulation of the SIR model
    for day in range(1, days):
        new_infected = beta * S[-1] * I[-1] / N
        new_recovered = gamma * I[-1]
        new_deaths = mu * I[-1]

        S.append(S[-1] - new_infected)
        I.append(I[-1] + new_infected - new_recovered - new_deaths)
        R.append(R[-1] + new_recovered)
        D.append(D[-1] + new_deaths)

    return S, I, R, D

# Initial parameters
N = 100000  # Total population size
beta = 0.4  # Disease transmission rate
gamma = 0.05  # Recovery rate of the disease
mu = 0.01  # Mortality rate (0.1% of the infected individuals)
I0 = 100  # Initial number of infected individuals
R0 = 0  # Initial number of recovered individuals
days = 150  # Number of days to simulate

# Simulate the SIR model
susceptible, infected, recovered, deaths = SIR_model(N, beta, gamma, mu, I0, R0, days)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(susceptible, label='Susceptible')
plt.plot(infected, label='Infected')
plt.plot(recovered, label='Recovered')
plt.plot(deaths, label='Deceased')
plt.xlabel('Days')
plt.ylabel('Number of individuals')
plt.title('SIR Model of Disease Spread with Mortality Rate')
plt.legend()
plt.grid(True)
plt.show()
```
