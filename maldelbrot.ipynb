```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def mandelbrot(c, max_iter):
    """
    Calculate the number of iterations required to determine if a complex number 'c' belongs to the Mandelbrot set.

    Parameters:
    c (complex): The complex number to be evaluated.
    max_iter (int): The maximum number of iterations allowed to classify 'c' within the Mandelbrot set.

    Returns:
    int: The number of iterations needed to classify 'c' within the Mandelbrot set, 
         or 'max_iter' if the limit is not reached.
    """
    z = c
    for n in range(max_iter):
        if abs(z) > 2.0:
            return n
        z = z * z + c
    return max_iter

def plot_mandelbrot(xmin, xmax, ymin, ymax, width, height, max_iter):
    """
    Plot the Mandelbrot set within the specified rectangular region.

    Parameters:
    xmin (float): The minimum x-coordinate of the rectangular region.
    xmax (float): The maximum x-coordinate of the rectangular region.
    ymin (float): The minimum y-coordinate of the rectangular region.
    ymax (float): The maximum y-coordinate of the rectangular region.
    width (int): The width of the rectangular region in pixels.
    height (int): The height of the rectangular region in pixels.
    max_iter (int): The maximum number of iterations allowed to classify each pixel within the Mandelbrot set.

    Returns:
    None: The Mandelbrot set is plotted to the current figure.
    """
    x = np.linspace(xmin, xmax, width)
    y = np.linspace(ymin, ymax, height)
    cmap = "hot"
    fig, ax = plt.subplots()
    mask = np.zeros((height, width))
    for i in range(height):
        for j in range(width):
            c = complex(x[j], y[i])
            mask[i, j] = mandelbrot(c, max_iter)
    
    ax.imshow(mask, extent=(xmin, xmax, ymin, ymax), cmap=cmap)
    ax.set_aspect("equal")
    ax.set_xlabel("Real")
    ax.set_ylabel("Imaginary")
    ax.set_title(f"Mandelbrot Set (Max Iterations: {max_iter})")
    plt.show()

if __name__ == "__main__":
    plot_mandelbrot(-2.5, 1.5, -2, 2, 500, 500, 100)
```
